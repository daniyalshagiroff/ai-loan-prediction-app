```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
```

```python
data_train = pd.read_csv('../data/train.csv')
data_test = pd.read_csv('../data/test.csv')
data_train = data_train.drop(columns=['loan_id'])
data_test = data_test.drop(columns=['loan_id'])
```

```python
data_train.head()
```

```python
mean_values = data_train[['income_annum', 'loan_amount', 'residential_assets_value', 
    'commercial_assets_value', 'luxury_assets_value', 
    'bank_asset_value', 'Total assets']].mean()
print(mean_values)
```

```python
most_frequent_values = data_train[['no_of_dependents', 'education', 'self_employed']].mode().iloc[0]
print(most_frequent_values)
```

```python
import pickle

with open('../data/mean_and_frequent_values.pkl', 'wb') as file:
    pickle.dump({'mean_values': mean_values, 'most_frequent_values': most_frequent_values}, file)

print("Saved in '../data/mean_and_frequent_values.pkl'")
```

```python
y_train = data_train['loan_status']
X_train = data_train.drop(columns=['loan_status'])
```

```python
X_test = data_test.drop(columns=['loan_status'])
y_test = data_test['loan_status']
```

```python
print(y_train.value_counts(normalize=True))
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)
```

```python
print(X_scaled[:5])
```

```python
regression = LogisticRegression(max_iter=1000)
```

```python
param_grid = {
    'C': [0.1, 1, 10],  
    'penalty': ['l2'], 
    'solver': ['lbfgs'], 
    'max_iter': [200, 500, 1000]  
}
```

```python
gridsearch = GridSearchCV(regression, param_grid, cv=5, scoring='accuracy')
```

```python
gridsearch.fit(X_train, y_train)
print("Best parameters found: ", gridsearch.best_params_)
print("Best cross-validation score: ", gridsearch.best_score_)
```

```python
best_model = gridsearch.best_estimator_
best_model.fit(X_train, y_train)
X_test.head()
y_pred = best_model.predict(X_test)
```

```python
X_test.head()
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
# Вывод метрик
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("\nConfusion Matrix:\n", conf_matrix)
```
